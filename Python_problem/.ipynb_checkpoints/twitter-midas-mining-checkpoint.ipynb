```python
'''tweepy was picked as the library as it eases interface to the Twitter API and reduces
our burden to interface and call Twitter API Directly.
This script sets up a connection to the twitter API through tweepy and dumps the MIDAS twitter page
tweet data onto a JSONLine File.
'''
import tweepy
from tweepy import OAuthHandler, Cursor
import json

#After setting up the app, the necessary keys are generated and are needed to provide
# application access to Twitter on behalf the account
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
#  In order to authorise the app to access Twitter, OAuth interface is used
client = tweepy.API(auth)
```

```python
'''Next we start mining data from our Twitter page and dump it onto a JSONLine file.
Each line containg a line of JSON or a tweet.Which makes it easy to process as the whole JSON file
need not be parses every time and accounts for faster and efficient memory access.'''
user="midasIIITD"
fname="user_timeline_{}.jsonl".format(user)
# the extension for JSONLine file is jsonl

with open(fname, 'w') as f:
# Cursor interface is used iterate through different types of objects in a page which 
# in turn contains multiple items for a particular user's timeline
    for page in Cursor(client.user_timeline,screen_name=user,count=200).pages(2):
        for status in page:
            f.write(json.dumps(status._json)+"\n")
# The status variable is an instance of the Status() class, wrapper to access the data. 
# The JSON response from the Twitter API is available in the attribute _json
# which is JSON string and not a dictionary written onto the file opened.
```
